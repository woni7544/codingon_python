# 용접기 AI 데이터분석 보고서

### 문제해결 장애 요인
- 용접 자동화 공정이 아닌 작업자 기반으로 작업이 되고 있어 판 두께에 따른 용접 조건을 경험 및 직관과 육안에 의존하고 있어 인공지능 모델에 필요한 자료를 수집하는데 어려움 존재

- 용접의 전류, 가압력, 통전시간을 PLC와 연동하여 MES에 수집하고, 판의 두께는 ERP-MES의 데이터를 용접 작업 오더 데이터를 가지고 활용
- 검사는 별도 인원 투입으로 양품/불량품 라벨링

## 3단계 : 오토인코더 모델 구축

- input_size : 입력의 크기
- hidden_size : 신경망층의 크기
- output_size : 출력의 크기
- nn.Linear() : 완전연결망
- nn.RReLU() : 각 신경망 층들을 더하면서 신경망을 구축

```python
class AutoEncoder(nn.Module):
    def __init__(self,input_size,hidden_size,output_size):
        super(AutoEncoder,self).__init__()
        ##initialize
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        ## 오토인코더 구현
        self.AutoEncoder = nn.Sequential(
        ## 인코더 부분
        nn.Linear(input_size,hidden_size[0]),
        nn.RReLU(),
        nn.Linear(hidden_size[0], output_size),
        nn.RReLU(),
        ## 디코더 부분
        nn.Linear(output_size, hidden_size[0]),
        nn.RReLU(),
        nn.Linear(hidden_size[0], input_size)
        )
    def forward(self,inputs):
        output = self.AutoEncoder(inputs)
        
        return output
```

## 4단계 : Training / Test 데이터 분할

- 분할하지 않고 모든 데이터를 Training 시키면 overfitting이 발생하여 오히려 성능이 저하
- training되지 않은 데이터도 제대로 학습하는지 검증

```python
#scaled_data = scaled_data.astype(float)
train_data = torch.Tensor(scaled_data_1[:7200])
print(len(train_data))
test_data = torch.Tensor(scaled_data_1[7200:])
print(len(test_data))
```

## 5단계 : 파라미터 지정

```python
# 훈련 하이퍼파라미터
epoch = 50
batch_size = 64
learning_rate = 0.01
# 모델 하이퍼파라미터
input_size = len(train_data[0])
hidden_size = [3]
output_size = 2
# 손실함수 제곱근 오차 사용
criterion = nn.MSELoss()
# 매개변수 조정 방식
optimizer = torch.optim.Adam
# 오토인코더 정의
AutoEncoder = AutoEncoder(input_size, hidden_size, output_size)
```

## 6단계 : 오토인코더 Training 실행

```python
def train_net(AutoEncoder, data, criterion,
              epochs, lr_rate = 0.01):
    optim = optimizer(AutoEncoder.parameters(),
                      lr = lr_rate)
    data_iter = DataLoader(data,batch_size = batch_size,
                           shuffle = True)
    for epoch in range(1, epochs+1):
        running_loss = 0.0
        for x in data_iter:
            optim.zero_grad()
            output = AutoEncoder(x)
            loss = criterion(x,output)
            loss.backward()
            optim.step()
            running_loss += loss.item()
            
        print('epoch : {}, loss: {:2f}'.format(epoch,running_loss))    
    return AutoEncoder
```

```python
AutoEncoder = train_net(AutoEncoder, train_data, criterion,epoch,learning_rate)
```

## 7단계 : Threshold도출 후 결과분석

```python
train_loss_chart = []
for data in train_data:
    output = AutoEncoder(data)
    loss = criterion(output,data)
    train_loss_chart.append(loss.item())
    
threshold_1 = np.mean(train_loss_chart) + np.std(train_loss_chart)*30
print('Threshold_1:', threshold_1)
```

```python
test_loss_chart = []
for data in test_data:
    ouput = AutoEncoder(data)
    loss = criterion(output,data)
    test_loss_chart.append(loss.item())
    
outlier_1 = list(test_loss_chart >= threshold_1)
outlier_1.count(True)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(len(test_loss_chart)), test_loss_chart, label='Test Loss')
plt.axhline(y=threshold_1, color='r', linestyle='--', label='Threshold_1')

plt.xlabel('Test')
plt.ylabel('Loss')
plt.title('Threshold_1')
plt.text(0.5, 0.95, f'Outliers Count: {outlier_1.count(True)}', transform=plt.gca().transAxes)
plt.legend()
plt.show()
```
