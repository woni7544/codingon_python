```python
import pandas as pd
taxi = pd.read_csv('taxi.csv',encoding = 'cp949')
taxi.head()
```

```python
for index, row in taxi['위도']:
    latitude = row['위도']
    longitude = row['경도']
m = fl.Map(location = [latitude, longitude], zoom_start = 10,width = 700, height = 500)
fl.TileLayer('Stamen Terrain').add_to(m)
fl.LayerControl().add_to(m)
fl.Marker( [latitude, longitude], popup = 'cozy house', tooltip = 'click plz', \
          icon = fl.Icon(color = 'red',icon = 'fa-solid fa-wifi',prefix='fa-solid')).add_to(m)
taxi
```
