```python
# 4. Scatter Plots
```

```python
import seaborn as sns
import pandas as pd
```

```python
#1번
```

```python
candy = pd.read_csv('candy.csv', index_col = 'id')
candy.head()
```

```python
#3번
```

```python
sns.scatterplot(x = candy['sugarpercent'], y = candy['winpercent'])
```

```python
#4번
```

```python
sns.lmplot(x='sugarpercent',y='winpercent',data = candy )
```

```python
#5번
```

```python
sns.scatterplot(x='pricepercent',y='winpercent',hue='chocolate',data = candy)
```

```python
#6번
```

```python
sns.lmplot(x='pricepercent',y='winpercent',hue='chocolate',data = candy)
```

```python
#7번
```

```python
sns.swarmplot(x='chocolate',y='winpercent',data = candy)
```
