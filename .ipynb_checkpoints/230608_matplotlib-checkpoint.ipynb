```python
import matplotlib.pyplot as plt
```

```python
#선그래프
```

```python
plt.plot([1,2,3],[3,5,1],'rs--',label = 'graph1')
plt.plot([1,2,3],[2,3,4],'b*--',label = 'graph2')
plt.plot([1,2,3],[5,1.5,4.5],'g',label = 'graph3')
plt.axis([1,3,1,6]) #x와 y축 범위 지정
font1 = {'family':'fantasy','color':'red','weight':'normal','size':'xx-large'} #글자서식 지정
plt.xlabel('xlabel', labelpad = 30, loc = 'center',fontdict = font1 )
plt.ylabel('ylabel', labelpad = 30, loc = 'center',fontdict = font1 )
#범례표시
plt.legend(loc = 'best',ncol = 1, frameon = True, shadow = True)
#제목표시
plt.title('Title', loc = 'center', pad = 50)
#grid 지정
plt.grid(True, axis = 'x')
plt.show()
```

```python
#막대그래프
```

```python
plt.bar([0,1,2],[100,200,300], color = ['r','b','g'])
#눈금표시
plt.xticks([0,1,2],['a','b','c'])
plt.show()
```

```python
plt.barh([0,1,2],[100,200,300], color = ['r','b','g'])
plt.yticks([0,1,2],['a','b','c'])
plt.show()
```

```python
#산점도
```

```python
import numpy as np
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
color = np.random.rand(n)

plt.scatter(x,y, s = 30, cmap = 'Spectral', c = color, alpha = 0.9 )
plt.show()
```

```python
#원그래프
```

```python

ratio = [34,32,16,18]
labels = ['a','b','c','d']
wedgeprops = {'width':0.7,'edgecolor': 'k', 'linewidth' : 1}
plt.pie(ratio, labels = labels, autopct = '%.1f%%', explode = [0.05,0.05,0.05,0.05], \
        wedgeprops = wedgeprops )
plt.show()
```

```python
month = ['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]



plt.plot(month,sales_2019,label = '2019')
plt.plot(month,sales_2020,label = '2020')
plt.legend()
plt.title('Monthly Sales Comparison(2019-2020)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20,35,15,27,45]
plt.title('BarChart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid()
plt.ylim([0,50])
plt.bar(categories,data)
plt.xticks(rotation=45)
```

```python
#실습3
```

```python
a = np.linspace(0,4*np.pi)
plt.xlim(0,10)
plt.title('Sine & Cosine Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.plot(a,np.sin(a),'b', label = 'sin(x)')
plt.plot(a,np.cos(a),'r', label = 'cos(x)')
plt.legend(loc = 'lower left')
plt.show()
```
